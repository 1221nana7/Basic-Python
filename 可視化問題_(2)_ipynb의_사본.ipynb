## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
data2 = pd.read_csv('https://raw.githubusercontent.com/shinonome-inc/Basic-Python/master/colab/quizzes/matplotlib_quiz/scatter_data.csv?token=GHSAT0AAAAAAB3AJEKZO2YEG3J4PVHM5VESY7SIODQ')
class0 = data2[data2['class']==0]
class1 = data2[data2['class']==1]
class2 = data2[data2['class']==2]
plt.scatter(class0['length (cm)'], class0['width (cm)'], label='class0', color='red')
plt.scatter(class1['length (cm)'], class1['width (cm)'], label='class1', color='blue')
plt.scatter(class2['length (cm)'], class2['width (cm)'], label='class2', color='green')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.title('クラス0〜2のlengthとwidthに関する散布図')
plt.grid(True)
plt.legend()
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(2).png

```python
data3 = pd.read_csv('https://raw.githubusercontent.com/shinonome-inc/Basic-Python/master/colab/quizzes/matplotlib_quiz/line_data.csv?token=GHSAT0AAAAAAB3AJEKY4GRNL7YJDMONAPPSY7SJQZQ')
plt.figure(figsize=(6,9))
plt.subplot(2,1,1)
plt.plot(data3['期間'], data3['件数'], marker='o')
plt.xlabel('期間')
plt.ylabel('件数')
plt.title('期間ごとの件数の推移')
plt.xticks(rotation=45)
plt.grid(True)

plt.subplot(2,1,2)
plt.bar(data3['期間'], data3['費用'])
plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.title('期間ごとの費用の推移')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
